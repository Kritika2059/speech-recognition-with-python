### **For 1-Channel Audio**

```python
import wave
import matplotlib.pyplot as plt
import numpy as np
```

### **Audio Signal Parameters**
number of channels

*   number of channels
*   sample width

*   ramerate/ sample_rate/ sample_frequency: 44,100
*   number of frames

*   values of frames

```python
obj = wave.open("/content/sample_wave.wav", "rb")

print("number of channels",obj.getnchannels())
print("sample width",obj.getsampwidth())
print("frame rate",obj.getframerate())
print("number of frames",obj.getnframes())
print("parameters",obj.getparams)

t_audio =obj.getnframes()/ obj.getframerate()
print("audio time: ", t_audio)

frames = obj.readframes(-1)
print("type of frames: ", type(frames), type(frames[0]))
print("length of frame: ", len(frames)/ 2)

obj.close()

obj_new = wave.open("sample_new.wav", "wb")

obj_new.setnchannels(1)
obj_new.setsampwidth(2)
obj_new.setframerate(16000.0)

obj_new.writeframes(frames)

obj_new.close()
```

## **Plot Audio**

```python
obj_plot = wave.open("/content/sample_wave.wav", "rb")

sample_freq = obj_plot.getframerate()
n_samples = obj_plot.getnframes()
signal_wave = obj_plot.readframes(-1)

obj.close()

t_audio_plot = n_samples/ sample_freq

print(t_audio_plot)

signal_array = np.frombuffer(signal_wave, dtype=np.int16)

times = np.linspace(0, t_audio, num=n_samples)

plt.figure(figsize=(15,5))
plt.plot(times, signal_array)
plt.title("Audio Signal")
plt.ylabel("Signal wave")
plt.xlabel("Time (s)")
plt.xlim(0, t_audio)
plt.show()
```
